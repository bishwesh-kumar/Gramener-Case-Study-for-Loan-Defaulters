UNIVARIATE & SEGMENTED ANALYSIS

Before we start with the Univiariate & Segment Univariate analysis, we would like to category Salaries & Loan amounts into categories for easy understanding

```python
# simple density plot
plt.hist(loan['funded_amnt'],bins='auto')
plt.show()
```

From the above histogram, though bins are created we are unable to target the exact loan group hence loan groups are made into groups as below 

```python
#UDF & creating to categorize funded loan amount column.
def loan_band(a):
    if a >= 5000 and a <10000:
        return 'Band2'
    elif a >=10000 and a <15000:
        return 'Band3'
    elif a >=15000 and a <20000:
        return 'Band4'
    elif a >=20000 and a <25000 :
        return 'Band5'
    elif a >=25000 and a <30000:
        return 'Band6'
    elif a >=30000 and a <35000:
        return 'Band7'
    elif a>=35000:
        return 'Band8'
    else:
        return'Band1'
```

```python
loan['Funded_Amount_Band']=loan['funded_amnt'].apply(loan_band)
loan.head()
```

```python
#UDF & creating to categorize Annual salary column & analyse further
def sal_cat(a):
    if a > 50000 and a <=100000:
        return 'Medium'
    elif a >100000 and a <=150000:
        return 'Above Medium'
    elif a >150000 and a <=200000:
        return 'High'
    elif a >200000:
        return 'Above High'
    else:
        return'Low'
```

```python
loan['Annual_Inc_Category']=loan['annual_inc'].apply(sal_cat)
loan.head()
```

```python
#Rounding of interest percentages for easy categorisation & analysis
loan['Int_Rounded_Percent']=round(loan['int_rate_percentage'],0)
loan.head()
```

A low debt-to-income (DTI) ratio demonstrates a good balance between debt and income. Conversely, a high DTI can signal that an individual has too much debt for the amount of income he or she has

```python
#Rounding of DTI percentages for easy categorisation & analysis
loan['DTI_Rounded_Percent']=round(loan['dti'],0)
loan.head()
```

BI VARIATE ANAYLSIS

Before we move to Bivariate analysis, we would like to derive few fields such as monthly income from annual income 

```python
loan['Monthly_Income']= round(loan['annual_inc']/12,0)
```

Deriving new field from DTI, to compare EMI percentage to the incoming salary to see any patterns

```python
loan['DTI in Currency']= (loan['Monthly_Income'] * loan['dti'])/100
```

```python
loan['Monthly_Inc_Sav_AfterDTI']= loan['Monthly_Income'] - loan['DTI in Currency']
```

```python
loan['EMI%_In_Sal']= round((loan['installment']/loan['Monthly_Inc_Sav_AfterDTI'])*100,0)
```
